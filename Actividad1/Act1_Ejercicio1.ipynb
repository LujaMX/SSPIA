```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

from IPython import display
from time import sleep
```

```python
def plot_2d(f,x,xl,xu):
    plt.clf()
    X = np.arange(xl, xu, 0.25)
    Y = f(X)

    plt.plot(X,Y)
    plt.plot(x, f(x), marker="o", c='r', markersize=10)

    plt.title("Plot")
    plt.xlabel('x')
    plt.ylabel('f(x)')

    plt.show(block=False)
    plt.pause(.05)
```

```python
f = lambda x: (20 - 2 *x) * (20 - 2 * x) * x
fp = lambda x: 12*(x**2) - 160*x + 400
fpp = lambda x: 24 * x + 160
```

```python
xl = 0
xu = 10
xi = 9.5
```

```python
for i in range(20):
    display.display(plt.gcf())
    display.clear_output(wait=True)

    #plot_2d(f,xi,xl,xu)
    xi = xi - (fp(xi) / fpp(xi))

plot_2d(f,xi,xl,xu)

if fpp(xi) > 0:
    print("Mínimo en x= ", xi, " f(x)= ", f(xi))
else:
    print("Máximo en x= ", xi, " f(x)= ", f(xi))
```
